A total of 19 visualizations were used in the analysis to create an effective data analytics report.
- Vertical Bar Charts: 6.
- Horizontal Bar, Trend, Pie, KPI, and Scatterplots: 2 each.
- Correlation Heatmap, Treemap, and Donut Chart: 1 each.

Extracting useful insights and transforming raw data into visual charts to create an effective data analytics report..This helps <br> 
in making better e-commerce decisions because visualization clearly highlights key insights, supports informed choices and <br>
assists in planning future goals. It also makes presentations easier and allows non-technical stakeholders to understand <br>
their ongoing business. Overall, this improves the e-commerce business by helping teams make the right decisions.<br>

## Importing Matplotlib & Seaborn library
Used to create visualizations.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,7))
counts = df["Age_category"].value_counts()

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n({count})"
    return my_format

counts.plot(
    kind="pie",
    autopct=autopct_format(counts.values),
    startangle=90)

plt.title('Percentage and count of customers who purchased products based on their "Age_category".', fontsize=20)
plt.ylabel("")  
plt.legend(
    labels=counts.index,
    loc="upper left")
plt.show()
```

The Age column was categorized into.
- 15â€“19 years (Adolescents)
- 20â€“29 years (Young Adults)
- 30â€“39 years (Adults)
- 40â€“49 years (Middle-Aged Adults)
- 50â€“59 years (Older Adults) 
- Above 59 years (Senior Adults)

```python
import matplotlib.ticker as mtick
plt.figure(figsize=(12,6))

category_age = df.groupby('Age_category')['Total_hours_spent_on_app'].sum().reset_index()
category_age = category_age.sort_values('Total_hours_spent_on_app', ascending=False)

bars = sns.barplot(
    x='Age_category',
    y='Total_hours_spent_on_app',
    data=category_age,
    color='brown')

for i, row in enumerate(category_age.itertuples()):
    bars.text(
        i,                                     
        row.Total_hours_spent_on_app / 2,      
        f"{row.Total_hours_spent_on_app:,.0f}",
        ha='center', va='center',
        color='white',
        fontsize=17, fontweight='bold')

plt.ylabel("Total Hours Spent", fontsize=5)
plt.xlabel('')
plt.title('Total hours spent on the app across "Age_category", ranked from highest to lowest', fontsize=20)
plt.show()
```

Reports indicate that the age groups spending the most hours on the e-commerce app follow this order:<br>
Adults > Middle-aged Adults > Older Adults > Young Adults > Adolescents > Senior Adults.<br>
This same ranking is observed in product purchases across age groups, suggesting that spending more <br>
time on the app motivates customers to buy more products.

```python
import matplotlib.ticker as mtick

sub_category_quantity = (
    df.groupby('Sub_category')['Quantity']
    .sum()
    .reset_index()
)

# Sort in descending order (high quantity first)
sub_category_quantity = sub_category_quantity.sort_values('Quantity', ascending=False)

plt.figure(figsize=(14,7))

bars = sns.barplot(
    y='Sub_category',
    x='Quantity',
    data=sub_category_quantity,
    color='orange'
)

# Format numbers
bars.xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Add labels inside bars
for i, row in enumerate(sub_category_quantity.itertuples()):
    bars.text(
        row.Quantity / 2,
        i,
        f"{row.Quantity:,.0f}",
        ha='center',
        va='center',
        color='black',
        fontsize=10
    )

plt.xlabel("Total Quantity Sold", fontsize=12)
plt.ylabel("")
plt.title('Total quantity sold of products across "Sub-Category", ranked from highest to lowest', fontsize=20)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as mtick

df["Delivered_date"] = pd.to_datetime(df["Delivered_date"])
df["Year"] = df["Delivered_date"].dt.year
trend_df = df.groupby("Year")[["Profit"]].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=trend_df, x="Year", y="Profit", marker="o", label="Profit")

for i, row in trend_df.iterrows():
    plt.text(
        row["Year"],
        row["Profit"],
        f"â‚¹{row['Profit']:,.0f}",
        ha="center",
        va="bottom",
        fontsize=11 )
  
plt.gca().yaxis.set_major_formatter(
    mtick.StrMethodFormatter('â‚¹{x:,.0f}'))

plt.title("Yearly Profit trend analysis",fontsize=20)
plt.xlabel("Year")
plt.ylabel("Total Profit")
plt.grid(True)
plt.legend()
plt.show()
```

The profit trend is analyzed from early 2019 to early 2025. The profit appears very low in 2025 because the <br>
dataset includes only the profit from the first half of January.

The years are arranged based on the ranking of total profit:
1. 2023 year
2. 2020 year
3. 2024 year
4. 2022 year
5. 2019 year
6. 2021 year
7. 2025 year (profit from only half of January)<br>

There is strong consistency in the profit range, which remains between â‚¹33,500,000 and â‚¹38,000,000 across most years.<br>
This indicates that the profit for the full year of 2025 is likely to fall between â‚¹33,500,000 and â‚¹38,000,000.
