# E-COMMERCE & SALES

## Introduction
E-commerce delivered sales records from 2019 to 2025 across various states in India, for customers aged 15 to 61,<br>
consisting of 40,000 rows and 25 columns of uncleaned and messy data.

## Objective
Clean and preprocess the raw dataset, analyze the sales and performance of e-commerce from 2019 to 2025 across ,<br>
various states in India, and develop a comprehensive visual data analytics report based on sales data.<br>
The analysis aims to answer key business questions essential for understanding overall e-commerce performance <br>
and customer purchasing behaviour.
- Which product is sold the most ?
- Which product generates the highest sales and profit ?
- Which region contributes the highest profit ? 
- What are the net profit and turnover of the business ?

## New column "Age_category" 

A new column was created based on the "Age" column to support better data analysis.
- Adolescents (15â€“19 years)
- Young Adults (20â€“29 years)
- Adults (30â€“39 years)
- Middle-Aged Adults (40â€“49 years)
- Older Adults (50â€“59 years)
- Senior Adults (above 59 years)

```python
def categorize_age(age):
    if 15 <= age <= 19:
        return "Adolescents"
    elif 20 <= age <= 29:
        return "Young adults"
    elif 30 <= age <= 39:
        return "Adults"
    elif 40 <= age <= 49:
        return "Middle-aged adults"
    elif 50 <= age <= 59:
        return "Older adults"
    else:
        return "Senior adults"

df["Age_category"] = df["Age"].apply(categorize_age)
```




Logical Imputation:<br>
Missing or null values were filled based on logical relationships among related columns.<br>
Since 'Numberof_ timesOrdered' = 'NumberofTimes_ordersAccepted' + 'Number_ofTimes_ordersDeclined',<br>logical imputation was applied to ensure data consistency.

```python
# Fill missing 'Numberof_ timesOrdered'
mask = (
    df['Numberof_ timesOrdered'].isna() &
    df['NumberofTimes_ordersAccepted'].notna() &
    df['Number_ofTimes_ordersDeclined'].notna()
)
df.loc[mask, 'Numberof_ timesOrdered'] = (
    df.loc[mask, 'NumberofTimes_ordersAccepted'] + df.loc[mask, 'Number_ofTimes_ordersDeclined']
)

# Fill missing 'NumberofTimes_ordersAccepted'
mask = (
    df['NumberofTimes_ordersAccepted'].isna() &
    df['Numberof_ timesOrdered'].notna() &
    df['Number_ofTimes_ordersDeclined'].notna()
)
df.loc[mask, 'NumberofTimes_ordersAccepted'] = (
    df.loc[mask, 'Numberof_ timesOrdered'] - df.loc[mask, 'Number_ofTimes_ordersDeclined']
)

# Fill missing 'Number_ofTimes_ordersDeclined'
mask = (
    df['Number_ofTimes_ordersDeclined'].isna() &
    df['Numberof_ timesOrdered'].notna() &
    df['NumberofTimes_ordersAccepted'].notna()
)
df.loc[mask, 'Number_ofTimes_ordersDeclined'] = (
    df.loc[mask, 'Numberof_ timesOrdered'] - df.loc[mask, 'NumberofTimes_ordersAccepted']
)
```

```python
print("MISSING VALUES AFTER LOGICAL IMPUTATION:")
df[['Numberof_ timesOrdered', 'NumberofTimes_ordersAccepted', 'Number_ofTimes_ordersDeclined']].isna().sum()
```

â€œThe 'Numberof_timesOrdered' column shows a symmetric distribution with no outliers. Therefore, mean imputation was<br>
applied to fill missingÂ values.

```python
ordered_mean_val = df["Numberof_ timesOrdered"].mean()
print("Mean value of Numberof_ timesOrdered column:", ordered_mean_val)
rounded = np.round(ordered_mean_val)
df["Numberof_ timesOrdered"] = df["Numberof_ timesOrdered"].fillna(rounded)
```

â€œThe 'NumberofTimesordersAccepted' and 'NumberofTimes_ordersDeclined' columns exhibit skewed distributions with outliers.
Therefore, median imputation was applied to fill missing values forÂ robustness.

```python
accepted_median_val = df["NumberofTimes_ordersAccepted"].median()
print("Median value of NumberofTimes_ordersAccepted column:", accepted_median_val)
df["NumberofTimes_ordersAccepted"] = df["NumberofTimes_ordersAccepted"].fillna(accepted_median_val)
```

```python
declined_median_val = df["Number_ofTimes_ordersDeclined"].median()
print("Median value of Number_ofTimes_ordersDeclined column:", declined_median_val)
df["Number_ofTimes_ordersDeclined"] = df["Number_ofTimes_ordersDeclined"].fillna(declined_median_val)
```

To ensure data consistency, the data type of the "Number_of_timesOrdered", "Number_of_Times_orders_Accepted", "Number_ofTimes_orders_Declined" columns were converted to an float.

```python
df["Numberof_ timesOrdered"] = df["Numberof_ timesOrdered"].astype(int)
df["NumberofTimes_ordersAccepted"] = df["NumberofTimes_ordersAccepted"].astype(int)
df["Number_ofTimes_ordersDeclined"] = df["Number_ofTimes_ordersDeclined"].astype(int)
```

it was observed that some values in the "Profit" column were incorrect or invalid. Therefore, the existing "Profit" column <br>
was removed, and a new one was created with accurate results using the standard accounting formula:<br>
Profit = Revenue âˆ’ Cost.

```python
df = df.drop(columns="Profit")
```

```python
df["Profit"] = df["Revenue"] - df["Cost"]
```

To ensure data consistency, the "Profit" column was converted to the float data type

```python
df["Profit"] = df["Profit"].astype(float)
```
